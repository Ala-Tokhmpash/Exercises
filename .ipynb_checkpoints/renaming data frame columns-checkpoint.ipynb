```python
all_names = titanic.Name
```

```python
all_names
```

```python
names_df = all_names.str.split(pat="(Mr\.) ([A-Za-z]+)", expand=True)
names_df
```

```python
names_df.columns = ["women", "Mr", "fn", "ln"]
```

```python
names_df
```

```python
men_names = names_df.fn.dropna()
men_names
```

```python
name_freq = men_names.value_counts()
name_freq
```

```python
from collections import Counter
```

```python
Counter(men_names)
```

```python
plt.rcParams.update({'axes.titlesize': 'large'})
plt.rcParams.update({'axes.labelsize': 'large'})

plt.figure(figsize=(20,10))
plt.bar(name_freq.index[240:], name_freq.values[240:])
plt.xlabel('Name', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
```

```python
plt.figure(figsize=(20,10))

sns.barplot(name_freq.index[:20], name_freq.values[:20],  palette="rocket")
```
